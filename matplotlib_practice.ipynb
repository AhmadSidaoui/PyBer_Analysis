```python
# dependencies

%matplotlib inline

import matplotlib.pyplot as plt
```

```python
# Importing Data

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month. 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Plotting Using MATLAB method

plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Object oriented plotting (1)

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
# Object oriented plotting (2)

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
# Annotating the Lineplot MATLAB method

# Create the plot and add a label for the legend. 
plt.plot(x_axis, y_axis, label='Boston', color="blue", linewidth=2, marker = "*")
# Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45. 
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.grid()
plt.show()
```

```python
# Annotating the Line plot using OO methode

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, label = "Boston", marker = "D", color = "g", linewidth = 2)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,45)
ax.legend()
ax.grid()
```

```python
# Plotting a vertical Bar chart MATLAP method

plt.bar(x_axis, y_axis, color = "g", label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
plt.show()
```

```python
# Plotting Horizontal Bar chart MATLAB method

plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.legend()
plt.title("PyBer Fare by Month")
plt.ylabel('Date')
plt.xlabel("Fare($)")
plt.gca().invert_yaxis()  # change order from ascending to descending and vis-versa
plt.show()
```

```python
# Plotting a vertical Bar chart OO method

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "green", label = "Boston")
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Fares($)")
ax.set_xlabel("Date")
ax.legend()
plt.show()
```

```python
# Plotting a horizontal Bar chart OO method

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fares($)")
ax.set_ylabel("Date")
plt.gca().invert_yaxis()  # For inverting the order of the y axis, we use plt.gca().invert_yaxis()
ax.legend()
plt.show()
```

```python
# Creating a scatter plot MATLAB method

plt.plot(y_axis, x_axis, 'o', color = "red", label = "Chicago")
plt.xlim(0,45)
plt.ylabel("Date")
plt.xlabel("Fares($)")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```

```python
# Creating a Bubble Chart (MATLAB method)

plt.scatter(x_axis, y_axis, s = y_axis, color = "r", label = "Chicago")
plt.title("PyBer Fare by Month")
plt.ylabel("Fare($)")
plt.xlabel("Date")
plt.ylim(0,45)
plt.legend()
plt.show()
```

```python
# utilizing List Comprehension

plt.scatter(x_axis, y_axis, s = [ i*5 for i in y_axis])
plt.show()
```

```python
# Creating Bubble chart (OO method)

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [ i*5 for i in y_axis], facecolors = "skyblue", alpha = 0.8, edgecolors = "black",
           linewidth = 2, label = "Boston")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,45)
plt.gca().invert_yaxis()
```

```python
# Creating Pie charts (MATLAB method)

exploded_values = [0,0,0,0,0,0,0.2,0,0,0,0,0]

plt.subplots(figsize = (8,8))
plt.pie(y_axis, labels = x_axis, explode = exploded_values, autopct = "%.1f%%")
plt.show()
```

```python
# Creating Pie charts (OO method)

exploded_values = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]

fig, ax = plt.subplots(figsize = (8,8))
ax.pie(y_axis, labels = x_axis, explode = exploded_values, autopct = "%.1f%%", shadow = True,
      startangle = 90)
plt.gca().invert_xaxis()
plt.show()
```

```python
# Creating Line chart with Error Bars (MATLAB method)

import statistics

stdev = statistics.stdev(y_axis)

plt.errorbar(x_axis, y_axis, yerr = stdev, capsize = 3)
plt.show()
```

```python
# Creating Line chart with Error Bars (OO method)

import statistics

stdev = statistics.stdev(y_axis)


fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr = stdev, capsize = 3)
plt.show()
```

```python
# Creating Error bars for bar charts (OO method)

import statistics

stdev = statistics.stdev(y_axis)

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr = stdev, capsize = 3)
plt.show()
```

```python
# Changing major ticks (MATLAB method)

import numpy as np

plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0,51,step = 5))
plt.gca().invert_yaxis()
```

```python
# Add minor ticks & Changing major ticks (OO method)

from matplotlib.ticker import MultipleLocator
import numpy as np

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,51, step = 5))
plt.gca().invert_yaxis()

ax.xaxis.set_minor_locator(MultipleLocator(1))   # adding the minor ticks
```
